# Use NGD for classifying infection vs non-infection

```python
import requests
from bs4 import BeautifulSoup
import math
import sys
import pandas as pd
```

```python
headers = {
'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
'Accept-Encoding': 'none',
'Accept-Language': 'en-US,en;q=0.8',
'Connection': 'keep-alive'
}
```

```python
def number_of_results(text):
    # headers is used for google request API
    r = requests.get("https://www.google.com/search?q="+text.replace(" ","+"),params={"gl":"us"},headers=headers)
    
    
    soup = BeautifulSoup(r.text, "lxml")
    res = soup.find("div", {"id": "resultStats"})

    print(res.text)
    for t in res.text.split():
        try:
            number = float(t.replace(",",""))
            print("{} results for {}".format(number,text))
            return number
        except:
              pass

    raise Exception("Couldn't find a valid number of results on Google")
    
```

```python
N = 25270000000.0
N = math.log(N,2)

def normalized_google_distance(w1, w2):
    f_w1 = math.log(number_of_results(w1),2)
    f_w2 = math.log(number_of_results(w2),2)
    f_w1_w2 = math.log(number_of_results(w1+" "+w2),2)

    return (max(f_w1,f_w2) - f_w1_w2) / (N - min(f_w1,f_w2))
```

```python
normalized_google_distance("infection", "pneumonia")
```

```python
df = pd.read_csv("~/Projects/Haimovich_Pressors/collection/out/diagnosis.csv")
```

```python
infection_ngd_list =[]
for diagnosis in df['diagnosis']:
    try:
        infection_ngd_list += [normalized_google_distance("infection", diagnosis)]
    except:
        infection_ngd_list += [-1]
```

```python
df['infection_ngd'] = pd.Series(infection_ngd_list)
```
